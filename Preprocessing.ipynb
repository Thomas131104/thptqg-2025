## Tiền xử lí bảng

```python
score = df.columns.drop(labels = ["Mã môn ngoại ngữ", "SOBAODANH"])
df["Toán"] = df["Toán"].fillna(0)
df["Văn"] = df["Văn"].fillna(0)
df[score] = np.round(df[score].astype(np.float32), 2).astype(np.float16)
```

```python
df_new = df[["SOBAODANH", "Toán", "Văn", "Mã môn ngoại ngữ"]].copy()
df_new
```

```python
import pandas as pd
import numpy as np

# Danh sách môn phụ
mon_phu = df.columns.drop(["STT","SOBAODANH","Toán","Văn", "Mã môn ngoại ngữ"])


def processing_optional_subject(row):
    result = []
    for mon in mon_phu:
        if pd.notna(row[mon]):
            result.append({
                "SOBAODANH" : row["SOBAODANH"],
                "Môn phụ" : mon,
                "Điểm môn phụ" : row[mon]
            })

    return result

records = []
for idx, row in df.iterrows():
    records.extend(processing_optional_subject(row))

df_melt = pd.DataFrame(records)
df_melt["Điểm môn phụ"] = df_melt["Điểm môn phụ"].round(2)
df_melt
```

```python
df_new = df_new.merge(df_melt, on="SOBAODANH", how="left")
df_new
```

```python
df_new["Mã sở"] = (df_new["SOBAODANH"] // 1000000).astype(np.int8)
df_new["Số thứ tự tỉnh"] = (df_new["SOBAODANH"] % 1000000).astype(np.int32)
df_new.drop(columns = ["SOBAODANH"], inplace = True)
```
